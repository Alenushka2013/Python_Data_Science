**Задача 1**

Напишите программу, которая получает на ввод n строк, состоящих из двух слов разделенных пробелом, и выводит сгруппированные элементы по второму слову, и отсортированные по алфавиту. Разделите каждую группу при помощи строки <->.

```python
from itertools import groupby
from operator import itemgetter
n = int(input())
cities = []

for i in range(n):
    cities.append(input().split())

result = []
cities.sort(key=itemgetter(1,0))
for country, cities in groupby(cities, key=itemgetter(1)):
    result.append('\n'.join([f"{i[1]} {i[0]}" for i in cities]))
print("\n<->\n".join(result))
```

```python
#  решение со словарем
d = {}

for _ in range(int(input())):
    city, country = input().split()
    d.setdefault(country, []).append(city)

res = ['\n'.join(f'{co} {ct}' for ct in sorted(d[co])) for co in sorted(d)]
print(*res, sep='\n<->\n')

```

**bisect**

Этот модуль очень полезен, если нужно часто вставлять элементы в отсортированный список. При этом не требуется постоянно сортировать массив после вставки. Производительность вставки – (O(n)).

Плюсом так же является тот факт, что поиск элементов осуществляется за логарифмическое время (алгоритм бинарного поиска).

```python
import bisect

a = [1, 3, 4, 4, 4, 6, 7]
# Найти место для вставки справа
print(bisect.bisect(a, 4)) # 5

# Найти место для вставки слева
print(bisect.bisect_left(a, 4)) # 2

# вставить элемент
bisect.insort(a,5)
print(a) # [1, 3, 4, 4, 4, 5, 6, 7]
```

**Задача 2**

Сначала вашей программе на вход поступает поток чисел. Сохраните их в массиве в отсортированном виде. После ключевого слова "change" меняется состояние программы, и на каждое введенное число она должна выдавать индекс элемента, на котором находится данное число (правый). Программа завершается, когда встречает слово "close".

```python
import bisect

a = []

while (n := input()) != "change":
    bisect.insort(a, int(n))
while (n := input()) != "close":
    print(bisect.bisect(a, int(n))-1)
```

**Задача 3**

Напишите программу, которая будет заменять символы строки на другую раскладку клавиатуры. Программа должна работать в обе стороны.

```python
eng_layout = "qwertyuiop[]asdfghjkl;'zxcvbnm,.`"
rus_layout = "йцукенгшщзхъфывапролджэячсмитьбюё"
s = input()
if s[0] in eng_layout:
    table = str.maketrans(eng_layout, rus_layout)
else:
    table = str.maketrans(rus_layout, eng_layout)
print(s.translate(table))
```
