**Задача 1**

Прочитайте текст из файла file.txt. Найдите в тексте слово, которое повторяется чаще всего. Запишите в переменную word само слово и в переменную word_count количество его повторений. В тексте в качестве разделителей используются пробел, точка и запятая. Регистр учитывать нужно.

```python
file = open('file.txt', 'r')
text = file.read().lower().replace('.', ' ').replace(',', ' ').split()
file.close()

d = {}
for word in text:
    if word in d:
        d[word] += 1
    else:
        d[word] = 1

word, word_count = '', 0
for key, value in d.items():
    if value > word_count:
        word_count, word = value, key

print(F'word = {word}\nword_count = {word_count}')
```

```python
from collections import Counter

with open('file.txt') as f:
    word, word_count = Counter(f.read().replace(',', ' ').replace('.', ' ').split()).most_common(1)[0]

print(F'word = {word}\nword_count = {word_count}')
```

**Задача 2**

Проверьте существование файла *data.txt*. Если файл существует – создайте новый файл *out.txt* и запишите в него содержимое файла *data.txt*. Если не существует, создайте файл *data.txt* со строкой Третий фактор.

```python
import os.path
if os.path.isfile("data.txt"):
    with open('data.txt', 'r')as file:
        text = file.read()

    with open('out.txt', 'w') as output:
        output.write(text)
else:
    with open('data.txt', 'w') as output:
        output.write('Третий фактор')
```

```python
try:
    with open('data.txt','x') as f:
        f.write('Третий фактор')
except FileExistsError:
    with open('out.txt', 'w') as fout, open('data.txt') as fin:
        fout.write(fin.read())
```
