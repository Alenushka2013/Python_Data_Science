**Задача 1**

Вам на вход поступает две даты в формате ISO 8601. Поменяйте в них местами годы если это возможно. Например на вход дано 2 даты:

`2023-01-01`

`2020-12-12`

В качестве ответа программа должна вернуть:

`2020-01-01`

`2023-12-12`

В случае если это сделать невозможно выведите:    *Год изменить невозможно*

```python
from datetime import date

d1 = date.fromisoformat(input())
d2 = date.fromisoformat(input())

year1 = d1.year
year2 = d2.year

try:
    d1 = d1.replace(year=year2)
    print(d1)

    d2 = d2.replace(year=year1)
    print(d2)

except ValueError:
    print('Год изменить невозможно')
```

Вашей программе на вход идут два числа – первый и второй год. Посчитайте количество високосных годов между ними включительно.

```python
n = int(input())
k = int(input())
count = 0
for i in range(n, k + 1):
    if i % 4 == 0 and i % 400 != 0:
        count += 1
print(count)
```

**Задача 3**

Вам дано 2 даты:* start_date* и *end_date*, где *start_date* <= *end_date*. Посчитайте, сколько между этими датами дней (включая даты), причём если *end_date* переходит на следующий месяц или ещё дальше – посчитайте, сколько дней между *start_date* и последним днём её месяца.

```python
from datetime import date
import calendar

d1 = date.fromisoformat(input())
d2 = date.fromisoformat(input())

year1 = d1.year
year2 = d2.year

month1 = d1.month
month2 = d2.month

day1 = d1.day
day2 = d2.day

if month1 == month2 and year1 == year2:
    print(day2 - day1 + 1)
else:
    print(calendar.monthrange(year=year1, month=month1)[1] - day1 + 1)
```

```python
from datetime import date
import calendar
d1, d2 = [date.fromisoformat(input()) for _ in range(2)]
d2 = min(d2, d1.replace(day=calendar.monthrange(d1.year, d1.month)[1]))
print((d2-d1).days + 1)
```

**Задача 5**

На вход вашей программы подается две строки t1  и t2  со временем в формате ISO 8601. Посчитайте, сколько времени прошло между t1  и t2.

 Время дается в разном порядке в рамках одного дня.

```python
from datetime import time
from datetime import timedelta

time1 = time.fromisoformat(input())
time2 = time.fromisoformat(input())

a = timedelta(hours = time1.hour, minutes = time1.minute, seconds = time1.second)
b = timedelta(hours = time2.hour, minutes = time2.minute, seconds = time2.second)

if time1 > time2:
    delta = a - b
else:
    delta = b - a

h, m, s = map(int, str(delta).split(':'))
time(hour=h, minute=m, second=s)
print(time(hour=h, minute=m, second=s))
```

**Задача 7**

На вход вашей программе поступают строка с датой, временем и часовым поясом в формате ISO 8601. Переведите время в часовой пояс UTC-03:46 и "Europe/Moscow" и выведите каждое время в новой строке.

```python
pip install pytz
```

```python
from datetime import datetime
import pytz

# Входные данные
sample_input = "2022-01-01T03:44:56+04:40"

# Создаем объект datetime из входной строки
dt = datetime.fromisoformat(sample_input)

# Переводим время в часовой пояс UTC-03:46
tz_utc_0346 = pytz.FixedOffset(-226)  # -03:46 = -3 часа 46 минут в минусе от UTC
dt_utc_0346 = dt.astimezone(tz_utc_0346)

print(dt_utc_0346.isoformat())

# Переводим время в часовой пояс "Europe/Moscow"
tz_moscow = pytz.timezone('Europe/Moscow')
dt_moscow = dt.astimezone(tz_moscow)

print(dt_moscow.isoformat())
```

```python
from zoneinfo import ZoneInfo
from datetime import datetime, timezone, timedelta

tz1 = timezone(offset=-timedelta(hours=3, minutes=46))
tz2 = ZoneInfo("Europe/Moscow")

d = datetime.fromisoformat(input())

print(d.astimezone(tz1).isoformat())
print(d.astimezone(tz2).isoformat())
```
