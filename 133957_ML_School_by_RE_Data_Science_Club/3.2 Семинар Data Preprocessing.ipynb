# Препроцессинг данных


**Вступление**
В этой статье мы поговорим о этапе который предшествует написанию любой модели машинного обучения - препроцессинг данных.

О препроцессинге данных
В реальном мире данные чаще всего искаженны. У них могут отсутствовать некоторые значения (или даже большинство), может столбик со строками быть заполнен числовыми данными, а может появиться машина, стоимостью в миллиарды. В процессе предобработки данных они обрабатываются и подготавливаются к обучению на них модели. В результате, мы получаем соответствующие требованиям определенной задачи данные.

**Кто этим занимается?**
Процессами обработки и хранения данных занимается Data Engineer. Их работа заключается в том, чтобы разбирать огромное количество данных в крупных компаниях. Data Scientist, Data Analyst и другие айтишники смотрят уже на красивые данные, которые обрабатывет DE.

**Зачем нужно обрабатывать данные?**
Существует такой принцип Garbage in - Garbage out (GIGO). Суть его в том, что - если данные, поступившие в модель были замусоренные, значит адекватного результата ожидать тоже не стоит. Некоторые модели крайне чувствительны к качеству данных. В машинном обучении 90% времени в ML тратится на подготовку данных.

# Четыре основных шага препроцессинга данных

**Чистка данных**

“Мусор - это то, чего не должно быть в данных”. Но что такое мусор и как его почистить. Разделим его на 3 составляющие. Отсутствующие данные, зашумленные данные и выбросы.

**Отсутствующие данные** (Nan-ки).

 Такое может быть, в большинстве случаев, при работе с данными из реальной жизни. В соревнованиях (Kaggle) обычно пустые значения отсутствуют. Например мы проходим опрос населения об эффективности работы ЖКХ. В нем было поле: “оценка качества воды”. Некоторые люди могли не написать в него ничего, по разным причинам, а у нас из-за них пустые значения. Что нам делать?

Во-первых мы можем их проигнорировать. Если мы хотим просто быстро проанализировать данные. Но ни одна модель пустые значения не примет, а если примет, она их преобразует сама.

Во-вторых мы можем просто удалить строки с пропусками. При небольшом количестве строк легче сделать так. Это самый легкий и не очень хороший вариант: мы можем потерять много полезной информации. Вдруг у всех, кто не оставил свою оценку воды она ужасная или ее просто нет.

Теперь перейдем к самому интересному. Заполнение пустых значений. мы можем заполнить пустоты разными способами.  Самое простое, что можно придумать, это заполнить либо средним либо самым частым значением.  Есть также более сложные подходы, можно заполнить пустые значения на основе предсказаний классификационных или регрессионных моделей, то-есть обучить модель предсказывать не пропущенные данные в этой колонке и затем заполнить пропуски.

**Зашумленные данные (Noisy Data).**

Шум – случайное значение измеряемой переменной. Причины появления шумов следующие:
сбои при сборе данных
сбои при вводе данных
сбои при передаче данных
сбои при преобразовании данных
несогласованность имен данных
технологические лимиты (на размер записи и др.)
Как нам избавиться от шумных данных. Есть 3 способа с помощью биннинга, регрессии и классификации.

**Биннинг (binning)** Смысл этого метода заключается в том, что мы сначала сортируем набор данных, затем разделяем на равномощные поднаборы, затем сглаживание средним, медианой или граничными значениями поднабора.

Пример Биннинга:

Отсортированные данные:

4, 8, 9, 15, 21, 21, 24, 25, 26, 28, 29, 34

Разделение данных на группы с одинаковой частотой

4, 8, 9, 15

21, 21, 24, 25

26, 28, 29, 34

Сглаживание с помощью среднего арифметического группы

9, 9, 9, 9

23, 23, 23, 23

29, 29, 29, 29

Сглаживание с помощью граничных значений группы

4, 4, 4, 15

21, 21, 25, 25

26, 26, 26, 34

**Регрессия** Сглаживание путем подгонки данных под регрессионную функцию

**Кластеризация** Создание групп/кластеров из данных, имеющих схожие значения. Значения, которые не находятся в кластере, могут рассматриваться как зашумленные данные и могут быть удалены. Кластеризацию мы пройдем в данном курсе позже, пока можно запомнить что она может помочь.

**Объединение данных**

Интеграция данных - это один из этапов предварительной обработки данных, который используется для объединения данных, имеющихся в нескольких источниках, в одно более крупное хранилище данных. Например: у нас 24 сейсмографа в определенной большой территории. Мы хотим как-то использовать эти данные. Модели устройств могут быть разные, они будут выдавать значения в разном виде и нам придется объединять их.

Интеграция данных необходима, особенно когда мы стремимся решить реальный сценарий, такой как обнаружение наличия заболеваний на изображениях компьютерной томографии. Единственный вариант - объединить изображения с нескольких медицинских узлов для формирования более крупной базы данных.

Мы можем столкнуться с некоторыми проблемами при внедрении интеграции данных в качестве одного из этапов предварительной обработки данных:

Интеграция схемы и сопоставление объектов: Данные могут присутствовать в различных форматах и с атрибутами, которые могут вызвать трудности при интеграции данных.
Удаление избыточных атрибутов из всех источников данных.
Обнаружение и разрешение конфликтов значений данных.
Пример интеграции двух датасетов с разной информацией:

```python
import pandas as pd

data1 = pd.read_csv("data1.csv")
data2 = pd.read_csv("data2.csv")

merged_data = pd.merge(data1, data2, on='ID')
```

**Изменение данных**

Теперь нам нужно изменить данные. Мы делаем так, чтобы данные были одинаковые по типам, убираем некоторые ненужные символы и приводим данные к похожему виду. К этому пункту также относятся более сложные процессы, как стандартизация и нормализация. Их мы разберем подробнее.

*Нормализация*

Статистика говорит нам так. Нормализация данных — это получение датасета с нулевым значением среднего арифметического и стандартным отклонением, которое равно единице. Для этого нужно вычесть среднее арифметическое значение из каждого наблюдения и затем поделить то, что получилось, на стандартное отклонение.

Допустим когда у нас очень большое значение зарплаты, оно не сожмётся до значения 1. Оно будет близким к нему, например 2,2. Такой подход с расширением диапазона позволяет не потерять важную информацию

Пример нормализации на numpy:

```python
import numpy as np

data = [1, 2, 3, 4, 5, 6, 7, 8, 9]

normalized_data = (data - np.mean(data)) / np.std(data)
```

Этот подход лучше стандартизации, ведь мы сохраняем исходное распределение и не делаем жестких границ

**Стандартизация**

В машинном обучении стандартизацией называют метод предобработки числовых признаков в обучающих наборах данных с целью приведения их к некоторой общей шкале например от -1 до 1.

Если у нас есть набор данных со значениями в диапазоне от 0 до 100, стандартизация масштабирует эти значения так, чтобы они находились в диапазоне от -1 до 1. Это позволяет упростить сравнение данных, а также делает их более подходящими для определенных алгоритмов машинного обучения. Стандартизация может быть применена к различным типам данных, таким как числовые данные, категориальные данные или текстовые данные.

**Десятичное масштабирование (decimal scaling)**

В данном методе нормализация производится путём перемещения десятичной точки на число разрядов, соответствующее порядку наибольшего числа: x’i=xi/10n
 , где n - число разрядов в наибольшем из чисел в выборке.

Например, пусть имеется набор значений: -10, 201, 301, -401, 501, 601, 701. Поскольку n=3, то получим числа: 0.01, 0.201, 0.301, -0.401, 0.501, 0.601, 0.701.

**Сокращение объема данных**

В некоторых случаях вам понадобиться уменьшить количество ваших данных. Это делается чаще всего из за мультиколлинеарности. Этот термин обозначает высокую корреляцию между двумя колонками, из которой следует их не информативность. Модели в процессе обучения такие повторяющиеся колонки не помогут и их можно убрать из данных. Например размер ноги зависит от роста. Мы можем убрать одну из этих колонок, в данном случае лучше размер ноги, и тем самым облегчить модель, повысить скорость и не потерять качества предсказания.
