```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

data = load_iris()

KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(data.data, data.target)
```

Sklearn позволяет в несколько строчек реализовать все модели в данном курсе, но тем не менее нужно понимать их внутреннюю работу, не говоря уже о том, что возможно вам придется реализовать и изменить алгоритм для специфичной задачи. Что касается KNeighborsClassifier в этой библиотеке, то у этой реализации есть много интересных параметров, например можно задать разные веса для каждой точки, условно ее важность, или изменить сам алгоритм на работающий на деревьях.

Давайте разберемся что тут произошло. Сначала была импортирована модель, затем инициализирован класс и туда был параметр n_neighbours=3, это то же самое что и
k выше. А  затем модель была обучена на известных данных.

Теперь мы можем использовать ее и предсказать значение для нового цветка.

```python
KNN.predict([[0.7, 0.3, 0.9, 0.3]])
```
