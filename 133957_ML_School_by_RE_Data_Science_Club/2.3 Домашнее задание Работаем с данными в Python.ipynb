```python
import pandas as pd
import numpy as np

# Обработка входных данных
data = input()
data = np.array([int(i) for i in data.split(" ")])
n = int(len(data)**(0.5))
data = np.reshape(data, (n, n))
df = pd.DataFrame(data = data)

# Your code here
suma, suma1 = 0, 0
for i in range(len(df)):
    suma += df.loc[i, i]
    suma1 += df.loc[i, len(df) - i - 1]
print(suma, suma1)

# print(np.trace(data), np.trace(data[::-1]))
# print(sum(np.diagonal(df)), sum(np.diagonal(np.fliplr(df))))
# print(np.trace(data), np.trace(np.rot90(data)))
```

**Задача № 5 **.

 На вход подаются данные, из которых создается датафрейм такой структуры:

 image.png

 Посчитайте среднее значение цен на каждый вид фруктов. В результате должна быть таблица, имеющая 2 столбца, вид фрукта и средняя цена на этот вид.

```python
import pandas as pd

# Преобразование входных данных
data = [input() for i in range(3)]

fruits = data[0].split(" ")
rating = [float(s) for s in data[1].split(" ")]
prices = [int(s) for s in data[2].split(" ")]

# Создание датафрейма
df = pd.DataFrame({'fruit': fruits * 3,
                   'rating': rating,
                   'price': prices})

# Your code here
print(df.groupby('fruit', as_index = False).price.mean().round(2))
```

**Задача № 6**.

На входе формируется датафрейм такой структуры.

image.png

*Этап 1*. Преобразуйте данные, применив в каждом столбце minimax нормализацию. Также округлите каждое значение до двух знаков после запятой.

*Этап 2.* Добавьте новый стролбец с названием "mean". В этой строке должны быть средние значения(np.mean) расчитанные по каждой строке, округленные до 2 знака после запятой.

*Этап 3.* Теперь ваша задача добавить новую строку снизу с индексом "sum". В этой строке должны содержаться суммы(np.sum) каждого столбца (в том числе и столбца "mean"). Результат суммы также должен быть округлен до второго знака после запятой.



image (1).png

```python
import pandas as pd
import numpy as np

# Преобразование данных
data = np.array(input().split(" "))
data = data.reshape(10, 10)
data = data.astype("int")
df = pd.DataFrame(data)

# Your code here
def min_max_normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# Применяем нормализацию ко всем столбцам
df = df.apply(min_max_normalize).round(2)

# Вычисляем средние значения по каждой строке
df['mean'] = np.mean(df, axis=1).round(2)

# Добавляем новую строку "sum" с суммами каждого столбца
df.loc['sum'] = df.sum().round(2)

print(df)
```

**Задача № 7.**

 На входе формируется датафрейм, имеющий такую структуру:

image.png

 Ваша задача, в каждой колонке отдельно, заполнить пропуски средним(np.mean) по объектам. Другими словами, в колонке "value1", пропущенное значение у объекта 2, заполняется средним значением по остальным объектам 2 в этой колонке.

```python
import numpy as np
import pandas as pd

# Обработка входных данных
data = input()
data = data.split(" ")
data = np.array(data).reshape(20, 2)
df = pd.DataFrame(data=data, columns=["value1", "value2"], index=["obj_1", "obj_2"]*10)
df = df.applymap(lambda x: np.nan if x == "nan" else float(x))

# Your code here
v1o1 = df['value1']['obj_1'].mean().round(2)
v1o2 = df['value1']['obj_2'].mean().round(2)
v2o1 = df['value2']['obj_1'].mean().round(2)
v2o2 = df['value2']['obj_2'].mean().round(2)

s1o2 = df.value1['obj_2']
s1o2.fillna(v1o2, inplace = True)
s1o1 = df.value1['obj_1']
s1o1.fillna(v1o1, inplace = True)
s2o2 = df.value2['obj_2']
s2o2.fillna(v2o2, inplace = True)
s2o1 = df.value2['obj_1']
s2o1.fillna(v2o1, inplace = True)

df.value1['obj_2'] = s1o2
df.value1['obj_1'] = s1o1
df.value2['obj_2'] = s2o2
df.value2['obj_1'] = s2o1

df = df.reset_index()
df = df.set_index('index')

print(df)
```
