# Разработка проекта анализа больших данных

1. **Условие практического задания**

Имеется конкретная задача или проект анализа больших данных. Например, прогнозирование уровня урожайности зерновых или классификация сельскохозяйственных предприятий по уровню банкротства.

2. **Постановка задачи**

Требуется спланировать процесс конкретного проекта по анализу больших данных. Например: анализ состояния растительности (идентификация сорняков, классификация растительности, выявление болезни растений, прогнозирование урожайности), анализ состояния почвы (оценка состояния почвы, классификация почвы, прогнозирование плодородности почвы) и другие.

Процесс анализа больших данных обычно состоит из 6 шагов

image.png

*Назначение цели исследования*. Цель исследования зависит от масштаба проекта. Однако, ее четкая формулировка играет ключевую в роль в конечном успехе всего проекта.

*Сбор данных.* На данном этапе необходимо четко представлять какие данные вам понадобиться и где их найти. Необходимо проверить насколько реально найти данные для вашего проекта, каково их качество и доступность. Следует ли приобретать данные у третьих лиц (сторонних организаций) и в каком формате они будут представлены (таблицы Excel или разного рода базы данных). Источники данных могут быть внутренними или внешними, открытыми или платными.

*Подготовка данных.* Данный этап состоит из трех под этапов: очистка данных – удаление некорректных значений из источника данных и устранение расхождения между источниками; интеграция данных – расширение информации посредством объединения информации из нескольких источников; преобразование данных – подбор подходящего формата данных для возможности их использования в модели.

*Исследование данных.* Данный этап позволяет более глубоко понять данные: как переменные взаимодействуют друг с другом, как распределены данные, имеются ли выпадающие значения. Основными методами исследования здесь является: описательная статистика, методы визуализации и построение простых моделей.

*Моделирование данных и построение модели.* На данном этапе знания предметной области, информация о данных, полученные на предыдущих этапах позволяют найти ответы на вопросы всего исследования. Здесь применяются методы статистики, машинного обучения, исследования операций и др.

*Отображение и автоматизация.* Результаты исследования должны быть представлены заказчику, научному сообществу, бизнес стороне и так далее. Результаты могут быть продемонстрированы в виде презентации, отчета, веб-страницы и так далее. Иногда процесс анализа требуется автоматизировать, так как информация текущего проекта может быть использована в другом проекте или задействована в рабочем процессе.

Стоит отметить, что процесс анализа больших данных носит итеративный, а не линейный характер. Иными словами, вам придется периодически возвращается к тому или иному этапу анализа и что-то переделывать. Например, на этапе исследования данных обнаружатся выпадающие значения (выбросы), что во многом указывает на возможные ошибки при импортировании данных. Также при построении модели факторные переменные могут быть подобраны не в полной мере или они будут коллинеарны и всю модель придется перестраивать.

3.** Пример решения**

В качестве примера рассмотрим классическую задачу прогнозирования URL-адресов.

**Этап 1**: цель проекта – определение того, можно ли доверять некоторым URL-адресам или нет.

**Этап 2**: в настоящем примере данные могут быть загружены из открытых источников. Например, с сайта sysnet (http://sysnet.uscd.edu/projects/url/#datasets).

**Этап 3**: подготовка данных в настоящей задаче заключается в проблеме нехватки памяти и необходимости использования дополнительных инструментов для загрузки в память компьютера. То есть, третий этап в настоящем примере достаточно ограничен. Связано это с тем, что URL-адреса проходят предварительную очистку, а вот провести здесь предварительное исследование данных потребуется.

**Этап 4**: на данном этапе может быть проведена проверка на наличие нулей в исходных данных. При выявлении нулевых значений может быть проведено соответствующее преобразование.

**Этап 5**: в качестве модели выявления вредоносных файлов может быть использован классификатор статического градиентного спуска SGDClassifier, который позволяет с высоким уровнем вероятности определить является ли веб-сайт вредоносным или нет.

**Этап 6**: Отображение и автоматизация результатов здесь может быть представлена в виде простой презентации в PowerPoint или в виде простейшего веб-приложения для моделей машинного обучения. Для этого можно изучить библиотеку Python Streamlit.

Теперь выберите подходящий для вас проект и опишите каждый из представленных этапов.
