# Применение метода опорных векторов

1. **Условие практического задания**

Имеется данные по ценам на недвижимость в сельской местности (файл Rural_prices.csv), а также сведения характеризующие данные объекты недвижимости (размер, наличие газа, зоны отдыха, наличие асфальтированной дороги к дому и так далее).

2. **Постановка задачи**

Требуется построить модель регрессии методом опорных векторов, предсказать значения цен на тестовой выборке, оценить качество модели на основе средних ошибок и сделать выводы.

3. **Пример решения**

Метод опорных векторов (Support Vector Machine, SVM) - это алгоритм машинного обучения, который используется для задач классификации и регрессии. Он находит оптимальную разделяющую гиперплоскость в многомерном пространстве признаков, используя опорные векторы на краях областей принятия решений.

SVM способен обрабатывать данные с большим количеством признаков (в нашем случае это 12 признаков и 546 наблюдений) и может работать с нелинейными данными, используя ядра для преобразования исходных данных в более высокоразмерное пространство, где они становятся линейно-разделимыми.

Для построения модели методом опорных векторов (SVM) в Python, нам потребуется библиотека scikit-learn.

Пример кода для создания модели SVM для имеющихся данных.

```python
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Прежде всего необходимо загрузить имеющие данные.

```python
df = pd.read_csv('/content/Rural_prices.csv', sep = ';', decimal = ',')
```

```python
df
```

Далее построение модели будет состоять из следующих этапов, представленных в коде.

```python
#Выберем переменные для построения регрессии.
y = df[['sale price']]
X = df[['lot size','#bedroom', '#bath', '#stories', 'driveway', 'rec room', 'basement', 'gas', 'air cond', '#garage', 'desire loc']]

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Масштабирование данных
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Создание модели SVM

svc = SVC(kernel='linear', probability=True)
svc.fit(X_train, y_train)
```

Теперь у нас есть модель SVM, обученная на наборе данных Rural_prices.csv. Мы можем использовать эту модель для предсказания цен на недвижимость, используя функцию predict() следующим образом:

Этот код создаст модель опорных векторов и предскажет цену каждого объекта недвижимости в тестовой выборке.

```python
y_pred = svc.predict(X_test)
print(y_pred)
```

В результате модель предсказывает значения - цены на жилье в сельской местности по тестовой выборке - 10% от общего количества наблюдений.

Настоящую модель можно использовать для быстрой и достоверной оценки и прогнозирования цен на недвижимость в сельской местности.

Для оценки эффективности модели посчитаем среднюю ошибку в абсолютном и относительном выражении.

```python
#Преобразуем данные
y_pred=y_pred.reshape(55,1)

#Импортируем недостающие библиотеки
from sklearn.metrics import mean_absolute_error

#рассчитаем среднюю ошибку
print('Результаты регрессии методом опорных векторов')
print('MAE:', round(mean_absolute_error(y_test, y_pred)))
print('MAPE:',round(np.mean(np.abs((y_test - y_pred)/y_test))*100))
```

Величина средней ошибки в 25% позволяет считать настоящую модель пригодной для предсказания целевого показателя,  в данном случае цен на недвижимость в сельской местности.
