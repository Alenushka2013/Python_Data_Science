Скачайте датасет pima_indians_diabetes.csv. Импортируйте его при помощи pandas в ваш Jupyter Notebook.

````python
import pandas as pd
````

````python
df = pd.read_csv('/content/pima_indians_diabetes.csv')
````

````python
df.head()
````

````python
# Проверьте, есть ли пропуски в данных
df.isna().sum()
````

````python
# В качестве матрицы объект-признаки X выберите все столбцы кроме 'Outcome'
X = df.iloc[:, :-1]
````

````python
# В качестве целевой переменной y выберите столбец 'Outcome'
y = df['Outcome']
````

````python
# Разбейте данные на тренировочные и тестовые, установив обязательно stratify=y,
# test_size=0.2 и random_state=10 в методе train_test_split()
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.2, random_state = 10)
````

````python
# Обучите StandardScaler на тренировочных данных и стандартизируйте тренировочные данные, после чего на обученном стандартизаторе тестовые,
# как мы делали ранее в предыдущих шагах
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
````

````python
# Обучите модель логистической регрессии. Внутри модели установите параметры
#  max_iter=1000 и random_state=10
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_std, y_train)
````

````python
# Получите предсказания на тестовых данных в виде меток классов и вероятностей

df_coefs = pd.DataFrame(data=clf.coef_, columns=X.columns).T
df_coefs.columns = ['Coeffecient']
df_coefs
````

````python
y_pred = clf.predict(X_test_std)
y_pred_train = clf.predict(X_train_std)
````

````python
# Рассчитайте метрики Precision, Recall, F1 и ROC-AUC для тестовых данных,
# округлите значений метрик до одного знака после запятой (используйте метод round())

from sklearn.metrics import precision_score, recall_score

print(f'Precision Train: {precision_score(y_train, y_pred_train)}')
print(f'Precision Test: {round(precision_score(y_test, y_pred), 1)}')

print(f'Recall Train: {recall_score(y_train, y_pred_train)}')
print(f'Recall Test: {round(recall_score(y_test, y_pred), 1)}')
````

````python
from sklearn.metrics import f1_score

print(f'F1 Train: {f1_score(y_train, y_pred_train)}')
print(f'F1 Test: {round(f1_score(y_test, y_pred), 1)}')
````

````python
from sklearn.metrics import roc_auc_score

y_pred_score = clf.predict_proba(X_test_std)
y_pred_train_score = clf.predict_proba(X_train_std)

print(f'ROC-AUC Train: {roc_auc_score(y_train, y_pred_train_score[:, 1])}')
print(f'ROC-AUC Test: {round(roc_auc_score(y_test, y_pred_score[:, 1]), 1)}')
````
