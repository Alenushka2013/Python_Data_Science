```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression

df = pd.read_csv('pima_indians_diabetes.csv')
X = df.drop(['Outcome'], axis=1)
y = df.Outcome
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, stratify=y, random_state=10)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_valid = sc.transform(X_valid)
logit = LogisticRegression(random_state=10, max_iter=1000)
logit = logit.fit(X_train, y_train)
y_pred = logit.predict(X_valid)
y_pred_proba = logit.predict_proba(X_valid)[:, 1]
print(*[round(precision_score(y_valid, y_pred), 1), round(recall_score(y_valid, y_pred), 1),
        round(f1_score(y_valid, y_pred), 1), round(roc_auc_score(y_valid, y_pred_proba), 1)])
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

df = pd.read_csv("/content/pima_indians_diabetes.csv")
# Проверьте, есть ли пропуски в данных
# df.info()
df.describe()
df.isna().sum() / df.shape[0]
# В качестве матрицы объект-признаки X выберите все столбцы кроме 'Outcome'
X = df.drop('Outcome', axis=1)

# В качестве целевой переменной y выберите столбец 'Outcome'
y = df['Outcome']

# Разбейте данные на тренировочные и тестовые, установив обязательно stratify=y, test_size=0.2 и random_state=10 в методе train_test_split()
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    stratify=y,
                                                    test_size=0.2,
                                                    random_state=10)

# Обучите StandardScaler на тренировочных данных и стандартизируйте тренировочные данные, после чего на обученном стандартизаторе тестовые, как мы делали ранее в предыдущих шагах
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

# Обучите модель логистической регрессии. Внутри модели установите параметры max_iter=1000 и random_state=10
clf = LogisticRegression(max_iter=1000, random_state=10)
clf.fit(X_train_std, y_train)

# Получите предсказания на тестовых данных в виде меток классов и вероятностей
y_pred = clf.predict(X_test_std)
y_pred_train = clf.predict(X_train_std)
y_pred_proba = clf.predict_proba(X_test_std)
y_pred_train_proba = clf.predict_proba(X_train_std)

# Рассчитайте метрики Precision, Recall, F1 и ROC-AUC для тестовых данных, округлите значений метрик до одного знака после запятой (используйте метод round())
# В качестве ответа через пробел введите получившиеся значения метрик Precision, Recall, F1 и ROC-AUC соответственно
print(f'{round(precision_score(y_test, y_pred), 1)} {round(recall_score(y_test, y_pred), 1)} {round(f1_score(y_test, y_pred), 1)} {round(roc_auc_score(y_test, y_pred), 1)}')
```
