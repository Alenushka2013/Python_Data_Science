**Задача 1**

3.4 - 1.jpg

```python
import numpy as np

# количество точек
n = int(input())

#  координаты точeк
points = []
for _ in range(n):
    points.append(input().split())
points = np.array(points, dtype = float)

# координаты "средней" точки
x_avr = np.mean(points, 0)[0]
y_avr = np.mean(points, 0)[1]

# массив расстояний от средней точки до каждой точки массива points
dist = []
for i in range(n):
    dist.append(((x_avr - points[i, 0]) ** 2 + (y_avr - points[i, 1]) ** 2) ** 0.5)

print("центр в точке (%6.3f, %6.3f), r = %6.3f"%(x_avr, y_avr, max(dist)))
```

**Задача 2**

3.4 - 2.jpg

```python
import numpy as np

# Электрическая цепь состоит из n ветвей
# n = int(input())

#  Известны сопротивление резистора r каждой ветви  и сила тока в каждой ветви i
r = np.array(input().split(), dtype = float)
i = np.array(input().split(), dtype = float)

# эквивалентное сопротивление цепи
R = sum(1/r) ** (-1)

# силa тока в цепи
I = sum(i)

# напряжение в цепи
U = I * R

print("R = %6.3f Ом, I = %6.3f А, U = %6.3f В"%(R, I, U))
```

**Задача 3**

3.4 - 3.jpg

```python
import numpy as np

# угол наклона шланга в градусах
tilt_angle = np.array(input().split(), dtype = float)

# дальность полета воды в метрах
dist_water = np.array(input().split(), dtype = float)

# угол наклона
phi = float(input())

# коэффициенты линии тренда
a_poly = np.polyfit(tilt_angle, dist_water,2)

# дальность полета струи воды для произвольного угла
dist_phi = a_poly[0] * phi ** 2 + a_poly[1] * phi + a_poly[2]
print("Дальность: %6.2f м"%dist_phi)
```

```python
import numpy as np
import numpy.linalg as alg

a = np.array([[1, 1, 1], [-1, 3, 0], [0, -1, 3]])

b = np.array([2970, 540, 390])

inv_a = alg.inv(a)

x = np.dot(inv_a, b)

print("%4.0f р., %4.0f р., %4.0f р."%(x[0], x[1], x[2])) #  %tuple(x)
```
