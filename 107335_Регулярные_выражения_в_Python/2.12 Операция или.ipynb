1. Напишите регулярное выражение, которое найдёт в тексте следующие языки программирования: JavaScript, C++, Python.

`regex = r'JavaScript|C\+\+|Python'`

2. Найдите все слова да, нет, наверное в тексте. Они могут начинаться с букв разных регистров.

`regex = r'\b(?:да|нет|наверное|Да|Нет|Наверное)\b'`

3. Криптостример Григорий отмывает грязные деньги через миксеры, а также через пару своих кошельков. Но вот незадача, он потерял адрес своего биткоин кошелька, на который должны прийти все монеты. Помогите Грише. Напишите регулярное выражение, которое найдёт все адреса биткоин кошельков.

Адрес представляет собой набор из префикса (1, или 3, или bc1) и основной части длиной от 27 до 34 символов.

В основной части используются:

Весь латинский алфавит, кроме: O, I, l.
Все арабские цифры, кроме 0.

`regex = r"\b(?:1|3|bc1)[^ _0OIl\WА-Яа-яёЁ]{27,34}\b"`

4. Найдите трёхзначные номера, перед которыми идёт №, Номер, или Number.

`regex = r'(?:(?<=№ )|(?<=Номер )|(?<=Number ))\d{3}'`
