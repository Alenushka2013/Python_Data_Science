1. Напишите любое регулярное выражение, которое находит само себя полным совпадением.

`regex = r'^(.*)$'`

2. Напишите регулярное выражение, которое найдёт все слова «Ты» или «ты». Другие формы слова «ты», такие как «твой» и т. д. учитывать не следует.

`regex = r'\b[Тт]ы\b'`

3. Цвета в формате HEX - способ представления rgb-цветов в шестнадцатеричной системе счисления. Они состоят из 6 шестнадцатиричных чисел, каждая пара отвечает за свой цвет:

Две первые цифры — красный
Две в середине — зелёный
Две последние цифры — синий
Напишите регулярное выражение, которое будет находить валидные hex-цвета.

`regex = r'#[A-Fa-f0-9]{6}'`

4. Напишите регулярное выражение, которое найдёт все числа x, где:

`x ∈ [0, 1] т.е. 0 ≤ x ≤ 1`

`regex = r'(\b(?:0\.\d{1,2})\b|\b(?:[01](?!\.))\b)'`
`regex = r'0\.\d{1,2}\b|[0-1](?!\.)'`

5. Напишите регулярное выражение, которое найдёт все слова, содержащие в себе букву а

`regex = r'\b(?![А])[А-ЯЁёа-я]*[Аа][А-ЯЁёа-я]*\b'`

6. Напишите регулярное выражение, которое найдёт слова, после которых следующим символом идёт один из следующих знаков препинания: .,:?!;

`regex = r'[А-ЯЁёа-яA-Za-z]+(?=[.,:?!;])'`

Слова, состоящие из кириллических символов, но в них есть как минимум 1 некириллический символ

Слова, состоящие из некириллических символов, но в них есть как минимум 1 кириллический символ
