1. Напишите регулярное выражение, которое найдёт все повторяющиеся последовательности из трёх цифр, которые идут друг за другом. Для этого используйте скобочные группы.

`regex = r'(?P<name>[0-9]{3})(?P=name)'`


2. Напишите регулярное выражение, которое найдёт все слова и словосочетания, состоящие из двух одинаковых частей. Между одинаковыми половинами слова может стоять дефис.

`regex = r'\b(?P<name>[а-яё]+)-?(?P=name)\b'`


3. Напишите регулярное выражение, которое найдёт все повторяющиеся буквы в тексте.

`regex = r'(?P<name>[A-Za-zА-Яа-яёЁ])(?P=name)'`


4. Напишите регулярное выражение, которое найдёт все повторяющиеся последовательности из двух цифр, которые идут друг за другом. Используйте нумерованные группы.

`regex = r'([0-9]{2})\1'`

5. Напишите регулярное выражение, которое найдёт все последовательности if и <if>, но не <if и if>, стоящие между началом и концом строки.

`regex = r'^(<)?(?(1)if\>|if?)$'`
