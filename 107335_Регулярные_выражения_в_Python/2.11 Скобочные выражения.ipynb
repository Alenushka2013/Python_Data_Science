1. Напишите регулярное выражение, которое получит последовательность из любых символов от [^START] до {(END.)}.

`regex =  r'(?<=\[\^START\]).*(?=\{\(END\.\)\})'`

2. Напишите регулярное выражение, которое найдёт все символы /, слева и справа от которых ничего нет или стоят другие символы, не равные /.

`regex = r'(?<!/)/(?!/)'`

3. Напишите регулярное выражение, которое найдёт все последовательности x с чётной длиной.

`regex = r'(?:(?<=^)|(?<=[^x]))(?:xx)*(?=[^x]|$)'`

4. Напишите регулярное выражение, которое найдёт все имена и названия в тексте. Слова в начале предложения не считаются, так как невозможно проверить это имя или просто слово с заглавной буквы.

`regex = r'(?<![.?!]\s)(?<!\A)[А-ЯЁ][а-яё]+'`

5. Напишите регулярное выражение, которое найдёт все неотрицательные числа.

`regex = r'\b(?<!-)[0-9]{1,}\b'`
