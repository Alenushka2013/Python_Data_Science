```python
import numpy as np
import pandas as pd

movie = pd.read_csv('/content/movie_metadata.csv')
movie.head()

```

```python
genres = movie[['movie_title', 'genres']]
genres
for i in genres:
    print(i)
```

```python
for i in genres.columns:
  print(i)
```

```python
for i in genres.values:
    print(i)

for row in genres.values:
    for value in row:
        print(value)

```

```python
def reversator(value):
    return value[::-1]

for row in genres.values:
    for value in row:
        print(reversator(value))
```

```python
for row in genres.iterrows():
    print(row)

for i, row in genres.iterrows():
    print(row.map(reversator))

```

```python
for i, col in genres.iteritems():
    print(col)
```

```python
#%%timeit
for i, col in genres.iteritems():
    print(col.map(reversator))
```

```python

# Временные ряды
import numpy as np
import pandas as pd
%matplotlib inline

stock=pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv', index_col='Date', parse_dates=True)
```

```python

stock.head()
stock.index
stock['2010']
stock['2010-02':'2011-03']
stock.resample('2h').asfreq()
stock.resample('1w').mean()
stock.head(10)
stock.rolling(3)
stock.rolling(3).mean()
stock.rolling(3, min_periods=1).mean()
stock.expanding(3).mean()
stock.ewm(alpha=0.7).mean()
stock['Open'].plot()
ns=stock['Open'].rolling(10, min_periods=1).mean()
stock['Open'].plot()
ns.plot()
```
